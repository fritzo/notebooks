# fmod versus remainder

```python
from math import pi
import jax.numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4 * pi, 4 * pi, 200)
```

```python
fig, axes = plt.subplots(3, figsize=(8,8), dpi=200, sharex=True)
ax = axes[0]
ax.plot(x, np.arctan2(np.sin(x), np.cos(x)))
ax.set_ylabel("arctan2(sin(x),cos(x))")
ax = axes[2]
ax.plot(x, np.fmod(x + pi, 2 * pi) - pi)
ax.set_ylabel("fmod(x+pi,2*pi)-pi")
ax = axes[1]
ax.plot(x, np.remainder(x + pi, 2 * pi) - pi)
ax.set_ylabel("remainder(x+pi,2*pi)-pi")
for ax in axes:
    ax.set_yticks((-pi, 0, pi))
    ax.set_yticklabels(("-", "0", ""))
ax.set_xticks((-4 * pi, -2 * pi, 0, 2 * pi, 4 * pi))
ax.set_xticklabels(("-4", "-2", "0", "2", "4"))
plt.subplots_adjust(hspace=0.05)
```
