```python
import numpy as np
from scipy.stats import beta as Beta
from matplotlib import pyplot
%matplotlib inline
```

```python
p = np.arange(0.001, 1, 0.001)
u = np.log(p / (1-p))
du_dp = 1/p + 1/(1-p)
alpha_grid = [1.0, 5.0, 9.0, 0.1, 0.5, 0.9, 0.01, 0.05, 0.09]
beta_grid = [9.0, 5.0, 1.0, 0.9, 0.5, 0.1, 0.09, 0.05, 0.01]
fig, axes = pyplot.subplots(3, 3, figsize=(12,11))
fig.patch.set_facecolor('white')
for i, (ax, alpha, beta) in enumerate(zip(axes.reshape(-1), alpha_grid, beta_grid)):
    pdf = Beta.pdf(p, alpha, beta) / du_dp
    ax.plot(u, pdf, label='alpha={:0.2g}, beta={:0.2g}'.format(alpha, beta))
    ax.set_yscale('log')
    ax.legend()
    if i // 3 == 2:
        ax.set_xlabel('u = log(p/(1-p))')
    if i % 3 == 0:
        ax.set_ylabel('density of u')
    if i == 1:
        ax.set_title('Density of logit-Beta distribution')
pyplot.tight_layout()
pyplot.show()
```
