## Logistic versus SoftLaplace distributions

```python
import torch
import pyro.distributions as dist
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-5, 5, 200)
plt.plot(x, dist.Logistic(0, 1).log_prob(x), "k--", label="Logistic")
plt.plot(x, dist.SoftLaplace(0, 1).log_prob(x), "r-", label="SoftLaplace")
plt.plot(x, dist.Laplace(0, 1).log_prob(x), "g-", label="Laplace")
plt.legend(loc="best");
```
