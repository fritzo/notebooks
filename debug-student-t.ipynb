# Debugging Student's t distribution

See Alican's [pull request](https://github.com/probtorch/pytorch/pull/65).

```python
import numpy as np
import torch
from torch.autograd import Variable
from torch.distributions import StudentT
from scipy.stats import t
from matplotlib import pyplot
%matplotlib inline
```

```python
df, loc, scale = [torch.FloatTensor([x]) for x in [0.1, 0.0, 1.0]]
t_torch = StudentT(df, loc, scale).sample(sample_shape=(100000,)).view(-1).numpy()
t_numpy = t.rvs(df, loc, scale, size=100000)
t_torch = np.log(np.abs(t_torch))
t_numpy = np.log(np.abs(t_numpy))
t_torch.sort()
t_numpy.sort()
pyplot.figure(figsize=(8,8)).patch.set_color('white')
pyplot.scatter(t_torch, t_numpy, alpha=0.5)
pyplot.xlabel('pytorch')
pyplot.ylabel('numpy')
pyplot.title('scipy vs torch.FloatTensor')
```
