# Binomial likelihood

Can the Binomial likelihood be approximated as Gaussian in `logits`? This notebook aims to visualized that approximation.

```python
import torch
import pyro.distributions as dist
import matplotlib.pyplot as plt
```

```python
n = 100
ks = [1, 2, 5, 10, 20, 50, 80, 90, 95, 98, 99]
logits = torch.linspace(-10., 10., 300)
plt.figure(figsize=(8, 4), dpi=300).patch.set_color("white")
for k in [1, 2, 5, 10, 20, 50, 80, 90, 95, 98, 99]:
    k = torch.tensor(float(k))
    plt.plot(logits, dist.Binomial(n, logits=logits).log_prob(k))
plt.ylim(-10, 0)
plt.xlim(-10, 10)
plt.xlabel("logits")
plt.ylabel("log likelihood")
plt.title("Binomial({}, logits).log_prob(k) for k in {}".format(n, ks))
plt.tight_layout();
```

```python
n = 100
ks = [1, 2, 5, 10, 20, 50, 80, 90, 95, 98, 99]
logits = torch.linspace(-15., 15., 300)
plt.figure(figsize=(8, 4), dpi=300).patch.set_color("white")
for k in [1, 2, 5, 10, 20, 50, 80, 90, 95, 98, 99]:
    k = torch.tensor(float(k))
    x = logits
    plt.plot(x / (1 + x**2)**0.25, dist.Binomial(n, logits=logits).log_prob(k))
plt.ylim(-10, 0)
plt.xlabel("logits / (1 + logits**2)**0.25")
plt.ylabel("log likelihood")
plt.title("Binomial({}, logits).log_prob(k) for k in {}".format(n, ks))
plt.tight_layout();
```
